```python
import numpy as np
import matplotlib.pyplot as plt
# 데이터
years = np.array([2014,2015,2016,2017,2018,2019,2020,2021,2022,2023])
sales = np.array([120,135,160,180,175,200,220,210,230,250])

plt.figure(figsize=(4,3))
plt.plot(years, sales, linestyle='-', marker='o' ,color='r')
plt.title("대한독립만세")
plt.grid(1)
plt.show(1)
```

```python


data = np.random.normal(0,1,1000)

plt.figure(figsize=(4,3))
plt.hist(data, bins=30, density=True)
plt.title('히스토그램')
plt.show()

plt.figure(figsize=(4,3))
plt.hist(data, bins=30, density=False)
plt.title('히스토그램')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize=(7,7))

for i in range(9):
  plt.subplot(3,3,i+1)
  plt.plot(x, np.sin(x+i), label=f"sin(+{i})")
  plt.title(f"Graph {i+1}")
  plt.xticks([])
  plt.yticks([])
  plt.legend()

plt.tight_layout()
plt.show()
```

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/ai05/body.csv')
df
```

```python
df = pd.DataFrame(df)

plt.figure(figsize=(10,4))
columns_to_plot = ['height', 'weight']

for i, col in enumerate(columns_to_plot):
  plt.subplot(1,2,i+1)
  plt.hist(df[col], bins=30, color='steelblue', edgecolor='black')
  plt.title(f"{col}")
  plt.xlabel("value")
  plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))

plt.subplot(1,2,1)
plt.hist(df['height'], bins=30, color='steelblue', edgecolor='black')
plt.title('키 히스토그램')
plt.xlabel("value")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(df['weight'], bins=30, color='steelblue', edgecolor='black')
plt.title('몸무게 히스토그램')
plt.xlabel("value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.hist(df['height'], bins=20,edgecolor='white')
plt.title(f"키 히스토그램")
plt.xlabel("value")
plt.ylabel("Frequency")

plt.subplot(1,3,2)
plt.hist(df['weight'], bins=20, color='gray', edgecolor='white')
plt.title(f"몸무게 히스토그램")
plt.xlabel("value")
plt.ylabel("Frequency")

plt.subplot(1,3,3)
plt.scatter(df['height'],df['weight'],alpha=0.6,color='red')
plt.title(f"산점도")
plt.xlabel("value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
