# 이상치 확인
1. 먼저 사용하는 컬럼 booking_changes,previous_cancellations 에 대한 이상치를 찾아보았다

2. iqr이 0으로만 출력되어서 그래프를 활용해서 0의 비율이 얼마인지 시각으로 확인해보려고한다

```python
# 1. 이상치 - IQR 단일 변수
import numpy as np

bq1 = np.percentile(df['booking_changes'],25)
bq2 = np.percentile(df['booking_changes'],50)
bq3 = np.percentile(df['booking_changes'],75)
biqr = bq3 - bq1

print("Q1: ",bq1)
print("Q2: ",bq2)
print("Q3: ",bq3)
print("iqr: ",biqr)
```

```python
# 2. 파이차트로 확인해보기

plt.figure(figsize=(4,3))
plt.pie(df['booking_changes'],labels=df['booking_changes'])
plt.show()
```

```python
# 2 파이차트로 너무 오래걸려서 빈도수 확인을 위해 히스토그램으로 분포 확인하기
df['booking_changes'].value_counts()

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['booking_changes'])
plt.xlabel('booking_changes')
plt.ylabel('Frequency')
plt.show()
```

```python
# 박스플롯으로도 확인해보기

plt.figure(figsize=(6,4))
plt.boxplot(df['booking_changes'],vert=False)
plt.show()
```

# 먼저 이상치값들의 예약이 취소되었는지 확인해보기

```python
# 이상치 기준선 계산
lower_bound = bq1 - 1.5 * biqr
upper_bound = bq3 + 1.5 * biqr

# 이상치로 판단되는 값들을 조회
outliers = df[(df['booking_changes'] < lower_bound) | (df['booking_changes'] > upper_bound)]

print(outliers['booking_changes'].value_counts())
```

```python
# 취소율을 시각화하기

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# booking_changes 에 따른 취소율 계산
cancel_rate_by_chagnes = df.groupby('booking_changes')['is_canceled'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='booking_changes', y='is_canceled', data=cancel_rate_by_chagnes)
plt.title('Cancellation Rate by Booking Changes')
plt.xlabel('Booking Changes')
plt.ylabel('Cancellation Rate')
plt.show()
```

# 데이터 전처리가 덜 된거같아서 다시 전처리하기

```python
df['total_guests'] = df['adults'] + df['children'] + df['babies']
print(df['total_guests'])
```

```python
# 객실방에 몇명이 들어갈 수 있는가 호텔을 직접 조사할 수 있지만 수치로 계산하여 iqr값으로 이상치를 잡아내고 설정하기로 했습니다.

import numpy as np

tgq1 = np.percentile(df['total_guests'],25)
tgq2 = np.percentile(df['total_guests'],50)
tgq3 = np.percentile(df['total_guests'],75)
tgiqr = tgq3 - tgq1

print("Q1: ",tgq1)
print("Q2: ",tgq2)
print("Q3: ",tgq3)
print("iqr: ",tgiqr)
```

```python
# 박스플롯으로 이상치확인해보기
plt.figure(figsize=(6,4))
plt.grid(axis='x', linestyle='--') # x축에만 점선 그리드 표시
plt.boxplot(df['total_guests'],vert=False)
max_val = int(df['total_guests'].max())
plt.xticks(range(0,max_val + 1, 3))
plt.show()
```
