#  가설1. 예약 변경이 있으면 취소할 확률이 있는가?
1. 먼저 사용하는 컬럼 booking_changes,previous_cancellations 에 대한 이상치를 찾아보았다

2. iqr이 0으로만 출력되어서 그래프를 활용해서 0의 비율이 얼마인지 시각으로 확인해보려고한다

```python
# 1. 이상치 - IQR 단일 변수
import numpy as np

bq1 = np.percentile(df['booking_changes'],25)
bq2 = np.percentile(df['booking_changes'],50)
bq3 = np.percentile(df['booking_changes'],75)
biqr = bq3 - bq1

print("Q1: ",bq1)
print("Q2: ",bq2)
print("Q3: ",bq3)
print("iqr: ",biqr)
```

```python
# 2 파이차트로 너무 오래걸려서 빈도수 확인을 위해 히스토그램으로 분포 확인하기
df['booking_changes'].value_counts()

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['booking_changes'])
plt.xlabel('booking_changes')
plt.ylabel('Frequency')
plt.show()
```

```python
# 박스플롯으로도 확인해보기

plt.figure(figsize=(6,4))
plt.boxplot(df['booking_changes'],vert=False)
plt.show()
```

```python
# 이상치 기준선 계산
lower_bound = bq1 - 1.5 * biqr
upper_bound = bq3 + 1.5 * biqr

# 이상치로 판단되는 값들을 조회
outliers = df[(df['booking_changes'] < lower_bound) | (df['booking_changes'] > upper_bound)]

print(outliers['booking_changes'].value_counts())
```

```python
# 취소율을 시각화하기

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# booking_changes 에 따른 취소율 계산
cancel_rate_by_chagnes = df.groupby('booking_changes')['is_canceled'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='booking_changes', y='is_canceled', data=cancel_rate_by_chagnes)
plt.title('Cancellation Rate by Booking Changes')
plt.xlabel('Booking Changes')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
correlation = df['booking_changes'].corr(df['is_canceled'])
print(f'예약 변경 횟수와 예약 {correlation} ')
```

```python
# 객실에 부모와 어린이가 같이왔을때 자식들끼리 방을 쓰게하는경우는 있을거같은데 유아끼리만 방을 예약하진않을거같아서 결측치를 제거합니다
num1 = (df['adults'] + df['children'] == 0)
num2 = (df['babies'] >= 1)

unusual_guests = df[num1 & num2]
print(unusual_guests)

# 값이 나오지않아서 넘기기로합니다
```

# 가설3. 어른 2명이 예약햇을때 예약 취소가 많은가? (커플 여행일 경우가 있을 수 있으니)


```python
# 먼저 새로운 데이터를 비교해보기위해 어른두명이 예약한 표본과 그 외 표본으로 나눠서 비교해봤습니다

condition_two_adults = (df_cleaned['adults'] == 2) & (df_cleaned['children'] == 0) & (df_cleaned['babies']== 0)

df_two_adults = df_cleaned[condition_two_adults].copy()
df_other = df_cleaned[~condition_two_adults].copy()
```

```python
# 시각화를 하여 눈으로 확인해보기

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


canc_rate_two_adults =  df_two_adults['is_canceled'].mean()
canc_rate_other = df_other['is_canceled'].mean()

canc_data ={
    'title' : ['two_adults', 'other'],
    'cancel_rate' : [canc_rate_two_adults,canc_rate_other]
}

df_canc = pd.DataFrame(canc_data)

plt.figure(figsize=(4,3))
sns.barplot(x='title', y='cancel_rate', data=df_canc)

plt.title('two adults and other')
plt.xlabel('booking type')
plt.ylabel('cancel rate')
plt.ylim(0, max(df_canc['cancel_rate']) * 1.2)
plt.show()
```

시각적으로 보았을때 두 표본의 차이가 크지않아서 성인2명의 여행이 취소율이 높은가에 대해서는 채택하기 어렵다고 판단했습니다
