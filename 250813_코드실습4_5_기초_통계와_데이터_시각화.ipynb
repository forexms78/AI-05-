```python
import numpy as np

x = [5,7,8,9,10,12,13,14,20]

q1 = np.percentile(x,25)
q2 = np.percentile(x,50)
q3 = np.percentile(x,75)
iqr = q3 - q1


print(f"Q1: {q1}")
print(f"Q2: {q2} (중앙값)")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")

```

```python
# 박스 프롯 - 이상치 확인  (아주 교과서적인 방법이다)

plt.figure(figsize=(4,3))
plt.boxplot(x, vert=False) #vert = 가로로 출력
plt.grid()
plt.title("사분위수 시각화")
plt.show()
```

```python
# 이상치를 2개를 생성

x = [-100,5,7,8,9,10,12,13,14,20,79]

q1 = np.percentile(x,25)
q2 = np.percentile(x,50)
q3 = np.percentile(x,75)
iqr = q3 - q1

print(f"Q1: {q1}")
print(f"Q2: {q2} (중앙값)")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")

plt.figure(figsize=(4,3))
plt.boxplot(x, vert=False) #vert = 가로로 출력
plt.grid()
plt.title("사분위수 시각화")
plt.show()
```

```python
# 박스 프롯 - 커스터마이징 (1)

plt.figure(figsize=(6,4))
plt.boxplot(
    x,
    vert=False, #vert = 가로로 출력
    patch_artist=True, # 박스 색을 채워라
    boxprops=dict(facecolor = 'skyblue', color = 'blue'), #박스색을 채워라
    medianprops=dict(color = 'red', linewidth=3), #중앙갑 선스 타일
    whiskerprops=dict(color = 'gray', linestyle="--"),  #수염스타일
    capprops=dict(color = 'green'),
    flierprops=dict(marker='o',markersize=8, markerfacecolor='red', color='black') #이상값 스타일
    )
plt.grid(axis='x', linestyle='--',alpha=0.4)
plt.title("사분위수 시각화", fontsize=15, fontweight='bold')
plt.yticks([1],['Group A'])
plt.show()
```

```python
# 박스 프롯 - 커스터마이징 (2)

# 박스 스타일 딕셔너리 수정
box_style = dict(facecolor='skyblue', edgecolor='blue') # 'color' -> 'edgecolor'로 수정
median_style = dict(color='red', linewidth=3)
whisker_style = dict(color='gray', linestyle="--")
cap_style = dict(color='green')
flier_style = dict(marker='o', markersize=8, markerfacecolor='red', color='black')

plt.figure(figsize=(6, 4))
plt.boxplot(
    x,
    vert=False,
    patch_artist=True,
    boxprops=box_style,
    medianprops=median_style,
    whiskerprops=whisker_style,
    capprops=cap_style,
    flierprops=flier_style
)
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.title("사분위수 시각화", fontsize=15, fontweight='bold')
plt.yticks([1], ['Group A'])
plt.show()
```

```python
# 히스토그램 실습

weight = [68,81,64,56,78,74,61,77,66,68,59,71,80,59,67,81,69,73,69,74,70,65]
x= [50,66,60,53,70,76,80,90,100]

plt.figure(figsize=(8,4))

plt.hist(x, label='bins=10',alpha=0.7)
plt.hist(weight, label='bins=10-weight', alpha=0.7, color='red')

plt.legend()
plt.title("두개 히스토그램 예시")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 가상 데이터 생성 (정규 분포를 따르는 100개의 데이터)
data = np.random.normal(loc=10, scale=2, size=100)

# KDE 그래프 그리기
sns.kdeplot(data, fill=True, color='skyblue')

# 그래프 제목과 레이블 설정
plt.title('KDE (커널 밀도 추정) 예시')
plt.xlabel('값')
plt.ylabel('밀도')

# 그래프 표시
plt.show()
```
