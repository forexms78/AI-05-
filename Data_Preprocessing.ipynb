```python
# @title íŠ¹ìˆ˜ë¬¸ì ì œê±°í•˜ê¸°

import re

rtext = re.sub(r'[^\w\s]', '', text)    # ìˆ«ì(\w), ê³µë°±(\s)ì„ ì œì™¸í•œ ëª¨ë“  íŠ¹ìˆ˜ë¬¸ì ì œê±°
rtext = re.sub(r'[\n\t]', ' ', rtext)   # ì¤„ë°”ê¿ˆ(\n), íƒ­(\t) ì œê±°
rtext = re.sub(r'\s+', ' ', rtext)      # ì—°ì†ëœ ê³µë°±(\s+) ì œê±°
rtext
```

```python
# @title ë¶ˆìš©ì–´ ì œê±°

sptext = rtext.split(' ')
st_words = ['', 'ì§„ì§œ', 'ì•Šì§€ë§Œ']
nw = []

for w in sptext:
    if w not in st_words:
      nw.append(w)

rtext = ' '.join(nw)
rtext
```

```python
# @title DF ì‚¬ìš©ì‹œ applyë¡œ ì „ì²´ë°ì´í„° ì ìš©

import re

stopwords = ['', 'ì§„ì§œ', 'ì•Šì§€ë§Œ']

def cleaning(review):
  review = re.sub(r'[^/w/s]', '', review) # íŠ¹ìˆ˜ë¬¸ì
  review = re.sub(r'[\n\t]', ' ', review) # ì¤„ë°”ê¿ˆíƒ­
  review = re.sub(r'\s+',' ', review)     # ì—°ì†ëœ ê³µë°±
  review = review.strip()

  review = ' '.join([w for w in review.split(' ') if w not in stopwords])

  review = ' '.join([w for w in review.split(' ') if len(w) > 1])

  return review


df['clean'] = df['text'].apply(cleaning)
df[['text','clean']]
```

```python
# @title í…ìŠ¤íŠ¸ë­‰ì¹˜ë°ì´í„° -> í…ìŠ¤íŠ¸íŒŒì¼

import sentencepiece as spm
import pandas as pd
import re

df = pd.read_csv('nreview_mask.csv')

# ì¶”ê°€ ì“°ê¸°ëª¨ë“œë¡œ í…ìŠ¤íŠ¸ íŒŒì¼ ì—´ê¸°
with open('nreview_mask.txt', 'a', encoding='utf-8') as f:
  for text in df['text']:
    text = str(text)
    text = re.sub(r'[^\w\s]', '',text)
    text = re.sub(r'[\n\t]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()

    try:
      f.write(text+'\n')
    except:
      pass

```
